```python
import requests
from bs4 import BeautifulSoup
import os

# 定義了一個名為 download_img 的函數，用於從給定的 URL 下載圖片並將其保存到指定的路徑
def download_img(url,save_path):
    print(f"正在下載圖片:{url}")
    response = requests.get(url)
    # 'wb' 模式表示以二進制寫入模式打開文件，適用於寫入圖片等二進制數據。
    with open(save_path,'wb') as file:
        file.write(response.content)
        print("-" * 30)
    
def main():
    url="https://www.ptt.cc/bbs/Beauty/M.1703573530.A.73E.html"
    headers= {"Cookie":"over18=1"}
    # 取得網址
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text,"html.parser")
        # 相要抓取的 html tag
    spans = soup.find_all("span",class_="article-meta-value")
    # 取得陣列第3個
    title=spans[2].text
    # 儲存圖片的路徑
    dir_name = f"images/{title}"
    # 如果路徑不存在，才會建立資料夾
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
    links = soup.find_all("a")
    allow_file_name = ["jpg","png","jpeg","gif"]
    # 抓取所有的圖片連結
    for link in links:
        href = link.get("href")
        if not href:
            continue
    #    按點（.）分割成一個列表 或趣最後一個元素 並轉成小寫
        file_name = href.split("/")[-1]
        extension = href.split(".")[-1].lower()
        if extension in allow_file_name:
            print(f"檔案型態:{extension}")
            print(f"url:{href}")
            download_img(href,f"{dir_name}/{file_name}")
    #    print(extension)
        
if __name__ == "__main__":
    main() 
```
